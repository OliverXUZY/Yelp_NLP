## Retrieve topic text in subsentence

```python
target_food = ['taste', 'flavour', 'topping','Chocolate', 'strawberry', 'vanilla', 'mango', 'caramel', 'banana', 'coconut', 'raspberry']
```

```python
sub_food = list(filter(lambda x: True if any(term in sub.text[x] for term in target_food) else False, sub.index))
sub_food[:6]
```

```python
len(sub_food)
```

```python
food_sentence = sub.loc[sub_food,:]
food_sentence.head()
```

```python
food_sentence.to_csv(os.path.join(dir,'dataset','food_sentence.csv'),index = False)
```

```python
target_price = ['price', 'combo', 'charge']
sub_price = list(filter(lambda x: True if any(term in sub.text[x] for term in target_price) else False, sub.index))
len(sub_price)
```

```python
price_sentence = sub.loc[sub_price,:]
price_sentence.head(2)
```

```python
price_sentence.to_csv(os.path.join(dir,'dataset','price_sentence.csv'),index = False)
```
